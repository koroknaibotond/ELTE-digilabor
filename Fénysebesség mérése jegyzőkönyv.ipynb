### 2.2. Ábrázolás és a fénysebesség meghatározása - 6 pont

Ábrázolja a mért adatokat és illesztés segítségével határozza meg a fénysebességet empirikus értékét, illetve az elektronika saját késleltetését!

```python
def c(t,T0,egypercé):
    return T0 + egypercé*t
```

```python
x = array([0.98,22.53,28.626,34.722,52.113])
t = array([33.75,99,122.2,147,213])
```

```python
popt, pcov = curve_fit(c,x,t)
perr = sqrt(diag(pcov))

print(popt)
print(perr)
```

```python
figure(figsize = (10,8))
plot(x,t, ls = " ", marker = "d", markersize = 10, label = "rögzített adatok")
plot(x,c(x,*popt), label = "illesztett egyenes")
legend()
title("A fénysebesség meghatározása", size = 20)
xticks(size = 13)
yticks(size= 13)
xlabel("távolság [m]",size = 15)
ylabel("időkülönbség [ns]", size = 15)
```

#### 4.2.3. A modell alapján a közeg jellemzőinek meghatározása - 2 pont

Az illesztésnél használt modellfüggvény egyenlete **XXXXX**, amely alapján a hullámellenállás **XXX $\pm$ XXX _$\Omega$_** és a veszteségi tényező **XXX $\pm$ XXX** adódik.

```python
ohm2 = np.linspace(0,824,100)
```

```python
def MOD(R,Q,R0):
    return Q * ((R-R0)/(R+R0))
```

```python
plot(ohm2,MOD(ohm2,1,1))
```

```python
ohm=[0,10,43,94,314,824]
adatok=[-0.481,-0.331,-0.077,0.149,0.381,0.459]
```

```python
popt, pcov = curve_fit(MOD,ohm,adatok, p0 = [1,1])
perr = sqrt(diag(pcov))

print(popt)
print(perr)
```

```python
plot(ohm,adatok)
plot(ohm2,MOD(ohm2,*popt))
```

```python
figure(figsize=(10,8))
plt.subplot(2, 1, 1)
plot(ohm,adatok, ls = " ", marker = "d",label = "mért adatok")
plot(ohm2,MOD(ohm2,*popt), label = "illesztés")
xlabel(r'ellenállás [$\Omega$]', size = 15)
ylabel('arány', size = 15)
title('Hagyomásnyos plot', size =20)
legend()
xticks(size=13)
yticks(size=13)


plt.subplot(2, 1, 2)
semilogx(ohm,adatok, ls = " ", marker = "d", label = "mért adatok")
plot(ohm2,MOD(ohm2,*popt), label = "illesztés")
xlabel(r'ellenállás skálázva [$\Omega$]', size = 15)
ylabel('arányok', size = 15)
title('Logaritmizált plot', size = 20)
legend()
xticks(size=13)
yticks(size=13)

tight_layout()
```
